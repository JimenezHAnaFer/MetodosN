```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del modelo de Vasicek
alpha = 0.1
beta = 0.05
sigma = 0.03
r0 = 0.06  # Tasa de inters inicial
T = 1.0    # Horizonte temporal
N = 1000   # Nmero de pasos de tiempo
dt = T / N # Tamao del paso de tiempo

# Generar el proceso de Wiener (movimiento Browniano)
t = np.linspace(0, T, N+1)
dW = np.sqrt(dt) * np.random.randn(N)

# Simulacin del proceso de tasa de inters
r = np.zeros(N+1)
r[0] = r0
for i in range(1, N+1):
    r[i] = r[i-1] + alpha * (beta - r[i-1]) * dt + sigma * dW[i-1]

# Graficar la simulacin
plt.figure(figsize=(10, 6))
plt.plot(t, r, label='Tasa de inters simulada')
plt.xlabel('Tiempo')
plt.ylabel('Tasa de inters')
plt.title('Simulacin del modelo de Vasicek')
plt.legend()
plt.grid(True)
plt.show()
```
